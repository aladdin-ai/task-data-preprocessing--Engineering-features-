```python
import pandas as pd
import seaborn as sns
import user_agents
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
```

## Using visitor-interests.csv dataset, from the first 50 row extract Browser, OS, Device data from 'UserAgent' column and use it to create new columns for Browser, OS, Device

```python
visitor_interests = pd.read_csv("visitor-interests.csv")
visitor_interests
```

```python
visitor_interests = visitor_interests.loc[:50]
visitor_interests
```

```python
def extracting_Browser_OS_Device(UA):
    UA = user_agents.parse(UA)
    return pd.Series((UA.os.family , UA.browser.family , UA.device.family))
visitor_interests[["OS" , "Browser" , "Device"]] = visitor_interests["UserAgent"].apply(extracting_Browser_OS_Device)
visitor_interests
```
