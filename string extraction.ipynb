```python
import pandas as pd
import seaborn as sb
```

# Extracting Mr,Miss,Mrs,etc.. from 'Name' column and visualize it

```python
titanic = pd.read_csv("titanic_full.csv")
titanic
```

```python
def extracting_title(tit):
    tit1 = tit.split(",")[1]
    return tit1.split()[0]
titanic["Title"] = titanic["Name"].apply(extracting_title)
titanic
```

```python
titanic["Title"].value_counts()
```

```python
sb.countplot(data = titanic , y = "Title")
```

# Using sendy_logistics.csv extract user_id in a seperate column

```python
dt = pd.read_csv("sendy_logistics.csv")
dt
```

```python
def extracting_id(id):
    return id.split("_")[2]
dt["ID."] = dt["User Id"].apply(extracting_id)
dt
```
